# Modelling

```python
# importing libraries
import pandas as pd
import numpy as np
import re
import string
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, plot_roc_curve, roc_auc_score,confusion_matrix, plot_confusion_matrix, classification_report
```

#### Reading in csv files

```python
# reading in csv files
df_u = pd.read_csv('../data/unvectorized_gpu.csv')
# checking df
df_u.head()
```

```python
# dropping unnamed column
df_u.drop(labels=["Unnamed: 0"], axis=1, inplace=True)
```

```python
# checking df
df_u.head(2)
```

```python
#checking shape
df_u.shape
```

```python
# checking for corrupted values
df_u.isnull().sum().sort_values(ascending = False)
```

```python
# checking which rows are null
df_u[df_u["content_"].isna()]
```

```python
# dropping null values
df_u.dropna(inplace = True)
```

## 1. Setting X and Y values

```python
# filtering columns
features = [col for col in df_u._get_numeric_data().columns if col != 'subreddit']
```

```python
# assigning x and y values
X_u = df_u['content_']
y_u = df_u['subreddit']
```

```python
# Define training and testing sets.
X_train_u, X_test_u, y_train_u, y_test_u = train_test_split(X_u,
                                                    y_u,
                                                    test_size=0.25,
                                                    stratify=y_u,)
```

```python
# checking set shape
X_train_u.shape
```

```python
# checking set shape
y_train_u.shape
```

## 2. Baseline Accuracy

```python
# Baseline accuracy based on Nvidia
y_test_u.value_counts(normalize=True)
```

## 3. Classification

#### Multinomial Naive Bayes with Count Vectorizer on GridSearch

```python
#Setting up a pipeline up with two stages:
# 1. CountVectorizer (transformer)
# 2. Multinomial Naive Bayes (estimator)
# 3. Fitting in identified stop words in NLP

pipe = Pipeline([
    ('cvec', CountVectorizer(stop_words=['like', 'would', 'also', 'http', 'imgur', 'www',
                                         'pcpartpicker', 'reddit', 'youtube'])),
    ('nb', MultinomialNB())
])
```

#### Logistic Regression with Count Vectorizer on GridSearch

```python
# setting up 2nd pipeline for LogReg with count vec on gridsearch
pipe2 = Pipeline([
    ('cvec', CountVectorizer(stop_words=['like', 'would', 'also', 'http', 'imgur', 'www',
                                         'pcpartpicker', 'reddit', 'youtube'])),
    ('logreg_pipe', LogisticRegressionCV(max_iter=100000))
])
```

#### Count Vectorizer hyperparameter settings

```python
# setting up pipe parameters
# removing outliers with min_df and max_df
pipe_params = {
    'cvec__max_features': [500, 1_000, 2_000, 3_000, 4_000],
    'cvec__min_df': [1, 2],
    'cvec__max_df': [0.9, 0.95],
    'cvec__ngram_range': [(1,1), (1,2), (2,3)],
}
```

#### Multinomial Naive Bayes with TF - IDF on GridSearch

```python
# setting up 3rd pipeline for MNB with TF IDF on gridsearch
pipe3 = Pipeline([
    ('tvec', TfidfVectorizer(stop_words=['like', 'would', 'also', 'http', 'imgur', 'www', 
                                        'pcpartpicker', 'reddit', 'youtube'])),
    ('nb', MultinomialNB())
])
```

#### Logistic Regression with TF - IDF on GridSearch

```python
# setting up 4th pipeline for log reg with TF IDF on gridsearch
pipe4 = Pipeline([
    ('tvec', TfidfVectorizer(stop_words=['like', 'would', 'also', 'http', 'imgur', 'www', 
                                        'pcpartpicker', 'reddit', 'youtube'])),
    ('logregidf', LogisticRegressionCV(max_iter=100000))
])
```

#### TF - IDF hyperparameter settings

```python
# setting up pipe parameters
pipe_params_idf = {
    'tvec__max_features': [500, 1_000, 2_000, 3_000, 4_000, 5_000],
    'tvec__ngram_range': [(1,1), (1,2), (2,3)]
}
```

# WARNING
Running the models will take a long time, with models 2 & 4 taking more than 20 minutes each.

### 4. Model 1
#### Executing Gridsearch with Count Vectorizer on Multinomial Naive Bayes

```python
# Instantiate GridSearchCV

gs = GridSearchCV(pipe, # what object are we optimizing?
                  param_grid=pipe_params, # what parameters values are we searching?
                  cv=5) # 5-fold cross-validation.
```

```python
# training model
gs.fit(X_train_u, y_train_u)
```

```python
# printing cross val score
print(gs.best_score_)
```

```python
# printing train score
train_score_u = gs.score(X_train_u, y_train_u)
print("Train Score : ", train_score_u)
```

```python
# printing test score
test_score_u = gs.score(X_test_u, y_test_u)
print("Test Score : ", test_score_u)
```

```python
# getting the best parameters for classifier
gs.best_params_
```

```python
# overall classifier used
gs.best_estimator_
```

```python
# Get predictions
preds_u = gs.predict(X_test_u)
cmatrix_u = confusion_matrix(y_test_u, preds_u)
```

```python
# Visualization
print(cmatrix_u)
print(classification_report(y_test_u, preds_u))
tn, fp, fn, tp = confusion_matrix(y_test_u, preds_u).ravel()

sens = tp / (tp + fn)
spec = tn / (tn + fp)
prec = tp / (tp + fp)

modified_cm = []
for index,value in enumerate(cmatrix_u):
    if index == 0:
        modified_cm.append(['TN = ' + str(value[0]), 'FP = ' + str(value[1])])
    if index == 1:
        modified_cm.append(['FN = ' + str(value[0]), 'TP = ' + str(value[1])]) 

plt.figure(figsize=(10,10))
sns.heatmap(cmatrix_u, annot=np.array(modified_cm),
            fmt="",
            annot_kws={"size": 20},
            linewidths=.5,
            square = True,
            cmap = 'Blues',
            xticklabels = ['AMD', 'NVIDIA'],
            yticklabels = ['AMD', 'NVIDIA'],
            );

plt.ylabel('Actual label', fontsize = 17);
plt.xlabel('Predicted label', fontsize = 17);
plt.title('Accuracy: {:.1f}'.format(test_score_u)+'\n'
          +'Precision: {:.1f}'.format(prec)+'\n'
          +'Sensitivity: {:.1f}'.format(sens)+'\n', size = 17);
plt.tick_params(labelsize= 15)
```

```python
# Plotting ROC curve
plot_roc_curve(gs, X_test_u, y_test_u)
plt.plot([0, 1], [0, 1],
         label='baseline', linestyle='--')
plt.legend();
```

```python
# Printing ROC score
roc_auc_score(y_test_u, gs.predict_proba(X_test_u)[:,1])
```

# Creating Word Cloud Visualization

```python
# pulling out pipe contents
vectorizer = gs.best_estimator_.named_steps['cvec']
classifier = gs.best_estimator_.named_steps['nb']
```

```python
# checking MNB coefficients
classifier.coef_
```

```python
# Word Cloud Function
def fancycloud(targetimage, color, invert=False):
    # Create a dataframe containing the words and their coefficient
    wordcloudcontent = pd.DataFrame(data = [vectorizer.get_feature_names(),
                     classifier.coef_[0]]).T.rename(columns = {0 : 'Features', 1: 'Coef'}).sort_values(by = 'Coef', ascending = invert).reset_index(drop = True)
    if invert == True:
        wordcloudcontent['Coef'] = wordcloudcontent['Coef']*-1

    # Create a dictionary where the word is the key and the coefficient is the associated value   
    freq = {}
    for i,j in wordcloudcontent.values:
        freq[i] = j
    
    # Create a mask using an image (preferably a black and white vector graphic)
    mask = np.array(Image.open(targetimage))
    
    # Generate a wordcloud function
    wordcloud = WordCloud(max_font_size=100,
                      max_words=150,
                      background_color=None,
                      mode='RGBA',
                      colormap=color,
                      mask = mask)
    
    # Call the function using the frequencies
    wordcloud.generate_from_frequencies(frequencies=freq)
    
    plt.figure(figsize = (14,14))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.savefig('../images/wordcloud_pic.png')
    plt.show()
```

## 9. Technical Conclusion

Model 1 (Count Vectorizer with Multinomial Naive Bayes) provided the best accuracy score of **0.9 for train** and **0.89 for test**.
The parameters for the best model was the following:
cvec__max_features =: 4000
cvec__ngram_range : (1, 1)

Although all the other models provided higher accuracy train scores, their difference between train and test were also the largest, with logistic regression providing the largest score differences.

All in all, all models performed better than the **baseline accuracy of 0.5574**.

It can be seen in the final chart that the top 5 words to affect the classifier weights include card, driver, game, nvidia and amd.  

With graphics cards being the common variable, the next 2 important words: driver, game depict specific problems users have. This means that specific problems specified by users are critical to classifier accuracy.  

The next two words of amd and nvidia are also critical for classifier accuracy.

In order to improve the accuracy of the model, one step we can take is to remove http links themselves from the posts and tune hyperparameters with a larger range.  
We would also like to try other classifiers like SVM to see if accuracy can be better.
