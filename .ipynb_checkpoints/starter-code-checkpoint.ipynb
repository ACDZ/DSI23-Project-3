````python
# Import libaries
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
````

````python
url = "https://www.reddit.com/r/AskMen/"
response = requests.get(url)
soup = BeautifulSoup(response.text,'lxml')
````

Note: Your best bet is to create a list of dictionaries, one for each restaurant. Each dictionary contains the restaurant's name and path from the `href`. The result of your scrape should look something like this:

````python
rest_tables = soup.find("table", {'id':'restaurants'})
rest_tables
````

````python
restaurants = []

for rest_links in rest_tables.find_all("a"):
    restaurant = {}
    restaurant["name"] = rest_links.text
    restaurant["href"] = rest_links["href"]
    restaurants.append(restaurant)
````

````python
foods = []
for restaurant in restaurants:
    print("Scraping {}".format(restaurant['name']))
    href = restaurant['href']
    #this is an fstring technique
    #u get to run functions in the string in {}
    restaurant_url = f'https://pages.git.generalassemb.ly/rldaggie/for-scraping/{href}'
    restaurant_response = requests.get(restaurant_url)
    restaurant_soup = BeautifulSoup(restaurant_response.text,'lxml')
    table = restaurant_soup.find("table", class_ = "table")
    
    for row in table.find('tbody').find_all("tr"):
        
        cells = row.find_all('td')
        #see that data extracted out will be in a list
        #print(cells)
    
        food_deets = {}
        
        #taken from above
        food_deets['restaurant'] = restaurant['name']
        
        food_deets['name'] = cells[0].text
        food_deets['category'] = cells[1].text
        food_deets['calories'] = cells[2].text
        food_deets['fats'] = cells[3].text
        food_deets['carbs'] = cells[4].text
        
        foods.append(food_deets)
    
    #trottle for 3 seconds before next scrape
    time.sleep(3)
    
    
````
